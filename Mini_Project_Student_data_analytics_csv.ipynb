```python
from pyspark import SparkContext, SparkConf
import pandas as pd
conf = SparkConf().setAppName("Student mini project")
sc = SparkContext.getOrCreate(conf = conf)
```

```python
student_rdd = sc.textFile("/content/sample_data/student_data.csv")
student_rdd.take(3)
```

#Show the number of students in the file.

```python
header = student_rdd.first()
rdd = student_rdd.filter(lambda x : x!=header)
rdd.count()
```

```python
rdd = rdd.map(lambda x : x.split(","))
```

#Show the total marks archieved by Female and Male Students.

```python
rddWithGenderAndScore = student_rdd.filter(lambda x : x!=header).map( lambda x : ( (x.split(",")[1]) , int(x.split(",")[5])) )
rddWithGenderAndScore.collect()
```

```python
rddWithGenderAndScore.reduceByKey(lambda x,y : x+y ).collect()
```

#Show the total number of students that have  passed and failed.50+ marks are required to pass the course.

```python
rdd.map(lambda x : ("passed",1) if int(x[5])>50 else ("failed",1)).reduceByKey(lambda x,y : x+y).collect()

```

#Show the total number of students enrolled per course.

```python
rdd.map(lambda x : (x[3],1)).reduceByKey(lambda x, y : x+y).collect()
```

#Show the total marks that sudents have achieved per course.

```python
rdd.map(lambda x : (x[3],int(x[5]))).reduceByKey(lambda x, y : x+y).collect()
```

#Show the average marks that sudents have achieved per course.

```python
rdd.map( lambda x : (x[3], (int(x[5]), 1)) ).reduceByKey(lambda x,y : (x[0]+y[0],x[1]+y[1])).map(lambda x : (x[0],x[1][0]/x[1][1])).collect()
```
